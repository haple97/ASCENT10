```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
%matplotlib notebook
```

```python
# Read airport data
airport_data = pd.read_excel('./GHG and cost LC stages System Expansion.xlsx', sheet_name='Airport List Map')
```

```python
# Create an empty matrix with rows and columns being 257 airports on airport list 
demand_matrix = pd.DataFrame(columns=airport_data['Code'], index=airport_data['Code'])
demand_matrix
```

```python
# Turn airport names into a list of strings 
index_list = list(demand_matrix.columns)
```

```python
# Loop across rows (i) and columns (j), replacing the corresponding cell with name of route indicated by row-column
for i in range (len(demand_matrix)):
    for j in range (len(demand_matrix)):
        demand_matrix.iloc[i, j] = index_list[i] +'-' + index_list[j]
demand_matrix.to_csv('./demand_matrix.csv')
```

```python
for i in range (len(demand_matrix)):
    for j in range (len(demand_matrix)):
        route = demand_matrix.iloc[i,j]     # type: string
        location = demand_list.index[demand_list['DIRROUTE'] == route].tolist()
        if np.all(demand_list['DIRROUTE'].isin([route])): 
            demand_matrix.iloc[i, j] = demand_list.at[location[0], 'PASSENGERS']
        else: 
            demand_matrix.iloc[i, j] = 0

demand_matrix.to_csv('./demand_matrix_filled.csv')       
```
