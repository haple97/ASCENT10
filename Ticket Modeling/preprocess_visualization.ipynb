**DIFFERENCE IN FARE BETWEEN BUSINESS & ECONOMY VS. %ADDITIONAL DISTANCE**

1. 3 separate classes (F, B, E) vs 2 classes (F+B, E) --> Difference = [B-E] vs. [(F+B) - E]
2. %fare difference (instead of |fare difference|) vs %additional distance
3. Fare difference vs. number of coupons (also number of stops) - another equivalent metrics for time taken 

```python
OD_market = market2.groupby(['Dirroute','Origin', 'Dest', 'Class']).agg({'NonStopMiles':'mean',
                                                             'NonStopMiles_Bin':'mean',
                                                             'MktDistance':'mean',
                                                             'MktMilesFlown':'mean',
                                                             'Percentage Additional Distance':'mean',
                                                             'Pctg_Bin':'mean', 'MktFare':'mean',
                                                            'MktFare_Wins': 'mean'})

# OD_market = market2.groupby(['Dirroute','Origin', 'Dest', 'Class', 'MktCoupons']).agg({'NonStopMiles':'mean',
#                                                              'NonStopMiles_Bin':'mean',
#                                                              'MktDistance':'mean',
#                                                              'MktMilesFlown':'mean',
#                                                              'Percentage Additional Distance':'mean',
#                                                              'Pctg_Bin':'mean',
#                                                             'MktFare': 'mean', 'MktFare_Wins': 'mean'})
OD_market.reset_index(inplace = True)

```

```python
OD_market
```

```python
OD_market2 = OD_market.groupby(['Dirroute']).agg({'Class':'nunique'})
OD_market2.reset_index(inplace = True)
OD_market2 = OD_market2[OD_market2['Class'] >1]   # Class is now NUMBER of classes, not the Class label
OD_market2.reset_index(inplace = True, drop = True)
```

```python
OD_market2
```

```python
# Filter for directional routes with more than two classes
OD_market_filtered = OD_market[OD_market['Dirroute'].isin(OD_market2['Dirroute'].tolist())]
OD_market_filtered.reset_index(inplace = True, drop = True)
```

```python
max_val = OD_market_filtered['Percentage Additional Distance'].max()
min_val = OD_market_filtered['Percentage Additional Distance'].min()
print(max_val, min_val)

bins = np.arange(min_val, max_val + 0.000000000000001, 0.02)
OD_market_filtered['Pctg_Bin'] = np.digitize(OD_market_filtered['Percentage Additional Distance'], bins)
```

```python
# Obtain routes with 2 classes only
demo = market2.groupby(['Dirroute', 'Class']).agg({'NonStopMiles':'mean','NonStopMiles_Bin':'mean',
                                                   'Percentage Additional Distance':'mean',
                                                   'MktFare_Wins': 'mean', 'Passengers':'sum'})
demo.reset_index(inplace=True)

demo2 = demo.groupby(['Dirroute']).agg({'Class':'nunique'})
demo2.reset_index(inplace = True)
demot2 = demo2[demo2['Class'] >1]   # Class is now NUMBER of classes, not the Class label
demo.reset_index(inplace = True, drop = True)

# Filter for directional routes with more than two classes
demo_filtered = demo[demo['Dirroute'].isin(demo2['Dirroute'].tolist())]
demo_filtered.reset_index(inplace = True, drop = True)

max_val = demo_filtered['Percentage Additional Distance'].max()
min_val = demo_filtered['Percentage Additional Distance'].min()
print(max_val, min_val)

bins = np.arange(min_val, max_val + 0.000000000000001, 0.02)
demo_filtered['Pctg_Bin'] = np.digitize(demo_filtered['Percentage Additional Distance'], bins)
demo_filtered
```

```python
testt= demo_filtered.groupby(['NonStopMiles_Bin', 'Pctg_Bin', 'Class'])['Passengers'].sum().unstack()
testt.reset_index(inplace = True)
testt = testt.dropna()
testt.reset_index(inplace = True, drop = True)
testt
```

```python
data_subset = testt[testt['NonStopMiles_Bin'] == 1]
data_subset.reset_index(inplace = True, drop = True)
data_subset = data_subset[data_subset['Pctg_Bin'] < 20]
data_subset.reset_index(inplace = True, drop = True)

x = data_subset['Pctg_Bin'].tolist()

data_subset = data_subset.drop(columns = ['NonStopMiles_Bin', 'Pctg_Bin'])

# Make stacked plots with counts of passengers in 2 classes
plt.stackplot(x, [data_subset["Business"], data_subset["Economy"]], labels=['Business', 'Economy'])
plt.legend(loc='upper left')
plt.xlabel('Percentage of Additional Distance Group (2% each)')
plt.ylabel('Number of Passengers')
plt.show()

# Make stacked plots with percentage of passengers in 2 classes
# # Transform the data from raw data to percentage (fraction)
# data_perc = data_subset.divide(data_subset.sum(axis=1), axis=0)

# # Make the plot
# plt.stackplot(x, data_perc["Business"], data_perc["Economy"], labels=['Business', 'Economy'])
# plt.legend(loc='upper left')
# plt.margins(0,0)
# plt.xlabel('Percentage of Additional Distance Group (2% each)')
# plt.ylabel('Ratio')
# plt.show()

```
